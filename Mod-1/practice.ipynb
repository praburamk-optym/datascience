```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,12))

g=sns.lineplot(x='year', y='totalprod', data=df, hue='state', legend='full')
g.set(xlim=(1998,2012), ylim=(0,5e7), xticks=np.arange(1996,2013,1))
plt.legend(bbox_to_anchor=(1.05, 1))
```
